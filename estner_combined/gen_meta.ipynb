```python
import pandas as pd
from tqdm import tqdm
```

```python
with open('train_both.json', 'r') as f:
    text = f.read()
```

```python
levels = ['ner_1', 'ner_2', 'ner_3']
entity_categories = ['PER', 'LOC', 'GPE', 'ORG', 'PROD', 'EVENT', 'DATE', 'TIME', 'TITLE', 'MONEY', 'PERCENT',
                    'DOC_ORG', 'CARD', 'IBAN', 'DOC_PER', 'IDCODE', 'EMAIL', 'TEL']
```

```python
for level in levels:
    entities = []
    for elem in text.split(level + '": "')[1:]:
        elem = elem.split('"')[0]
        if elem[0]=='B':
            entities.append(elem.split('-')[1])
    print(level)
    print(pd.Series(entities).value_counts())
```
